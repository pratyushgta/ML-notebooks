## EXPERIMENT 8 - Task 1

Building a three-layer feed forward neural network from scratch by initializing random weights and biases.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
import numpy as np
```

```python
df= pd.read_csv("iris.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.shape
```

```python
df.dtypes
```

```python
numeric_columns = df.select_dtypes(exclude=["object"]).columns
categorical_columns = df.select_dtypes(include=["object"]).columns
```

```python
scaler = StandardScaler()

df[numeric_columns] = scaler.fit_transform(df[numeric_columns])
```

```python
df = pd.get_dummies(df, columns=categorical_columns, dtype=int)
```

```python
df.columns
```

```python
df.head()
```

```python
df.describe().round(2)
```

```python
input = df.iloc[:, 0:4]
```

```python
input.shape
```

```python
input
```

```python
np.random.seed(42)

w_i_h1 = np.random.rand(4,1)
w_i_h2 = np.random.rand(4,1)

w_h_o1 = np.random.rand(2,1)
w_h_o2 = np.random.rand(2,1)
w_h_o3 = np.random.rand(2,1)

bias1 = np.random.rand(1)
bias2 = np.random.rand(1)
```

```python
print(w_i_h1, '\n')
print(w_i_h2, '\n')
print('\n')
print(w_h_o1, '\n')
print(w_h_o2, '\n')
print(w_h_o3, '\n')
print('\n')
print(bias1, '\n')
print(bias2, '\n')
```

```python
h1_in = np.dot(input, w_i_h1) + bias1
h2_in = np.dot(input, w_i_h2) + bias2
```

```python
print(h1_in.shape)
print(h2_in.shape)
```

```python
def sigmoid(x):
  return 1/(1 + np.exp(-x))
```

```python
h1_out = sigmoid(h1_in)
h2_out = sigmoid(h2_in)
```

```python
print(h1_out.shape)
print(h2_out.shape)
```

```python
h_out = np.append(h1_out, h2_out, axis=1)
h_out.shape
```

```python
o1_in = np.dot(h_out, w_h_o1)
o2_in = np.dot(h_out, w_h_o2)
o3_in = np.dot(h_out, w_h_o3)
```

```python
o1 = sigmoid(o1_in)
o2 = sigmoid(o2_in)
o3 = sigmoid(o3_in)
```

```python
print(o1.shape)
print(o2.shape)
print(o3.shape)
```
