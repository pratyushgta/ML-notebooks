## EXPERIMENT 1 - Task 2

Perform one hot and label encoding on relationship column of “adults” dataset

```python
from sklearn import preprocessing
import pandas as pd
```

```python
# Label encoder is in-built class of sklearn library.
#We are creating object of this class called label_encoder
label_encoder = preprocessing.LabelEncoder()
```

```python
df = pd.read_csv("adult.data.csv")
df.columns
```

```python
df['relationship'].unique()
```

i. Label Encoding

```python
# converting data of 'relationship' column from non-numeric (categorical data) to numeric
# Performing Label Encoding
df['relationship_coded'] = label_encoder.fit_transform(df['relationship'])
# fit transform will transform the column to numeric value
```

```python
# label encoding converts each value in a column into a number
# gives random numbers to each value based on an in-built algorithm
# these numbers are just representations do not have an underlying meaning/ hierarchy
df
```

ii. One-hot Encoding

```python
# label encoding gives random numbers but one-hot encoding creates columns for each unique value and would work like adjacency matrix
df_new = pd.get_dummies(df,columns=['relationship'])
# to give more columns: df_new = pd.get_dummies(df,columns=['relationship']['col2']['col3'])
```
